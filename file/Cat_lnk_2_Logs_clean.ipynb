```python
import pandas as pd 
import os

def logs_file(path_file,input_folder):
    liste_des_fichiers = os.listdir(path_file)

    dico = {}
    for i in liste_des_fichiers:
        df = pd.read_csv(os.path.join('Logs',input_folder,i))
        #print(df)
        #df['type_location'] = 'Count_Send_Logs'
        dico[i[:10]] = df
        # concatenation des sheet
    concat_df = pd.concat(dico)
    #mettre les index en colonne
    concat_df = concat_df.reset_index()
    #suppression d'un index pas important
    concat_df = concat_df.drop(['level_1'], axis = 1)
    #rename de l'index qu'on veut sauvegarder
    concat_df = concat_df.rename(columns={"level_0": "date"})
    concat_df['date'] = pd.to_datetime(concat_df['date'], format='%d-%m-%Y')
    #export du nouveau fichier
    concat_df.to_excel(input_folder + '.xlsx', index = False)
    #print(concat_df)
```

# Probleme
1) Données NULL

On remarque qu'il y'a certains données manquantes concernant le tag_id, party_id et log_time.
Ceci peut se justifier sur le fait que dans certaines observations, on a des données decalées.

2) Données en doublant

On sait qu' une personne est identifiée par son tag_id donc ce dernier doit etre unique.
Mais pour cette table, nous avons 32 doublants par rapport à tag_id.

On remarque que certaines données sont doublées par rapport à la date,tag_id et party_id mais un log_time different.

3) Non Standardisation de la colonne log_time

log_time est sous le format : 
    * H:M
    * 1/0/00 H:M

# Solution 
1) Données NULL

Le probleme des données manquantes peut etre resolu en traitant le probleme des doublant

2) Données en doublant

Pour resoudre ce probleme, nous avons consideré qu'une personne emet qu'une seule fois un tag_id. Donc si un tag_id est detecté deux fois, nous allons garder le log_time le plus proche de l'heure de l'heure de Count and Send. En faisant un trie par ordre croissant de log_time, nous donneras toujours le premier des doublant qui a le log_time le plus petit.
