```python
import pandas as pd
# To summary, with the advanced method, the exercice is done in a few lines:
# Loading all excel sheets as a dic of df
dic_of_employee = pd.read_excel('Nettoye/Employee_Evening_Schedule_cor.xlsx', sheet_name=None)
# Adding the year columns to each df
for key, values in dic_of_employee.items():
    dic_of_employee[key]['date'] = key
# Concatenate all df
concat_df = pd.concat(list(dic_of_employee.values()))
concat_df
```

```python
df1 = pd.read_excel('Nettoye/24_05_22.xlsx')
df1['date'] = '24 05 2022'
df2 = pd.read_excel('Nettoye/13_06_22.xlsx')
df2['date'] = '13 06 2022'
df3 = pd.read_excel('Nettoye/17_06_22.xlsx')
df3['date'] = '17 06 2022'
df4 = pd.read_excel('Nettoye/05_06_22.xlsx')
df4['date'] = '05 06 2022'
```

```python
df4
```

```python
concat_employee = pd.concat([df1, df2, df3, df4, concat_df])
concat_employee
```

```python
concat_employee = concat_employee.drop(['Unnamed: 0','Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3'], axis = 1)
concat_employee
```

```python
concat_employee = concat_employee.drop(['Unnamed: 0'], axis = 1)
concat_employee
```

```python
concat_employee.info()
```

```python
concat_employee['date'] = pd.to_datetime(concat_employee['date'], format='%d %m %Y')
concat_employee
```

```python
concat_employee.to_csv('Evening_employee_schedule.csv', index = False)
```
